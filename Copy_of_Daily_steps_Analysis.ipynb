```python
 # import necessary libraries
import pandas as pd
import random
import matplotlib.pyplot as plt # import matplotlib

# Generate random steps data for one week
dates = pd.date_range(start="2025-01-15" , periods=7)
steps = [random.randint(3000, 10000) for _ in range(7)]

# Create a dataframe to store the data
df = pd.DataFrame({"Date": dates, "Steps": steps})

# Display the data
print("Steps for the week:")
print(df)

# Plot the steps data
df.plot(x="Date", y="Steps", kind="line", title="Steps for the week", grid=True)
plt.xlabel("Date")
plt.ylabel("Numbers of Steps")
plt.show()
```
